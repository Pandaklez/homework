```python
from pymystem3 import Mystem
from pymorphy2 import MorphAnalyzer
import csv
```

```python
morph = MorphAnalyzer()
m = Mystem()
```

```python
f = open('w.txt', 'r', encoding = 'utf-8')
s = f.readlines()
f.close()
```

```python
words = []
for el in s:
    #print(el)
    el = el.strip('\n\t\r')
    words.append(el)
print(words[:1038])
```

```python
def csv_dict_writer(fieldnames, data):
    with open("out.csv", "w", newline='') as out_file:
            writer = csv.DictWriter(out_file, delimiter=',', fieldnames=fieldnames)
            writer.writeheader()
            for row in data:
                writer.writerow(row)
    out_file.close()
```

```python
pos = []
z = []
anim = []
aspect = []
case = []
gender = []
involvement = []
mood = []
number = []
person = []
tense = []
transitivity = []
voice = []
for word in words:
    ana = morph.parse(word)[0]
    z.append(ana)
    pos.append(ana.tag.POS)
    anim.append(ana.tag.animacy)
    aspect.append(ana.tag.aspect)
    gender.append(ana.tag.gender)
    involvement.append(ana.tag.involvement)
    mood.append(ana.tag.mood)
    number.append(ana.tag.number)
    person.append(ana.tag.person)
    tense.append(ana.tag.tense)
    transitivity.append(ana.tag.transitivity)
    voice.append(ana.tag.voice)
    #d_inner[ana.word] = ana.tag
print(person)
```

```python
data = []
data.append(['lexeme','pos','anim','aspect','gender','involvement','mood','number','person','tense','transitivity','voice'])
k = len(z)
#i = 0
for i in range(0, k):
    data.append([z[i].normal_form,pos[i],anim[i],aspect[i],gender[i],involvement[i],mood[i],number[i],person[i],tense[i],transitivity[i],voice[i]])
    #i += 1
print(data)
```

```python
my_list = []
fieldnames = data[0]
for values in data[1:]:
    inner_dict = dict(zip(fieldnames, values))
    my_list.append(inner_dict)
csv_dict_writer(fieldnames, my_list)
```

```python
s = csv.DictReader(open("out.csv"))
cc = []
for row in s:
    content = row['pos']
    cc.append(content)
print(cc)
```
